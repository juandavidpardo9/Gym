```python
from abc import abstractmethod
class Calcular:
  @abstractmethod
  def cal(same):
    pass
  @abstractmethod
  def reiniciar(same):
    pass  
```

```python
class CalcularEstrato(Calcular):
  def iniciar(same):
    super().iniciar()
    same.estrato =0
  def cal_estrato(same,estrato) :
    e = int(estrato)
    if e >= 4 :
      same.estrato = 1
    elif e == 3 :
      same.estrato = 0.5 
    else :
      same.estrato = 0
  def reiniciar(same): 
    same.estrato = 0
  def cal(same):
    return (same.estrato)
class CalcularSituEconomica(Calcular):
  def iniciar(same):
    super().iniciar()
    same.sit_eco =0
  def cal_sit_eco(same,sit_eco) :
    if sit_eco == "Mejor":
      same.sit_eco = 1
    elif sit_eco == "Igual":
      same.sit_eco = 0.5
    else :
      same.sit_eco =0
  def reiniciar(same): 
    same.sit_eco = 0
  def cal(same):
    return (same.sit_eco)
class CalcularPersonasHogar(Calcular):
  def iniciar(same):
    super().iniciar()
    same.per_hogar =0 
  def cal_per_hogar(same,per_hogar,sit_eco):
    e = int(per_hogar)
    if e >= 5 and (sit_eco == "Mejor" or sit_eco == "Igual"):
      same.per_hogar = 1
    elif e == 3 and (sit_eco == "Mejor" or sit_eco == "Igual"):
      same.per_hogar = 0.5
    else:
      same.per_hogar = 0
  def reiniciar(same): 
    same.per_hogar = 0
  def cal(same):
    return (same.per_hogar)
class CalcularEduPadre(Calcular):
  def iniciar(same):
    super().iniciar()
    same.edu_padre=0
  def cal_edu_padre(same,edu_padre):
    if edu_padre == "Educaci—n profesional completa" or edu_padre == "Postgrado" or edu_padre == "TŽcnica o tecnol—gica completa" :
      same.edu_padre = 1
    elif edu_padre == "Educaci—n profesional incompleta" or edu_padre == "TŽcnica o tecnol—gica incompleta" :
      same.edu_padre = 0.5
    else :
      same.edu_padre = 0
  def reiniciar(same):
    same.edu_padre = 0
  def cal(same):
    return (same.edu_padre)
class CalcularEduMadre(Calcular):
  def iniciar(same):
    super().iniciar()
    same.edu_madre=0
  def cal_edu_madre(same,edu_madre):
    if edu_madre == "Educaci—n profesional completa" or edu_madre == "Postgrado" or edu_madre == "TŽcnica o tecnol—gica completa" :
      same.edu_madre = 1
    elif edu_madre == "Educaci—n profesional incompleta" or edu_madre == "TŽcnica o tecnol—gica incompleta" :
      same.edu_madre = 0.5
    else :
      same.edu_madre = 0
  def reiniciar(same):
    same.edu_madre = 0
  def cal(same):
    return (same.edu_madre)
class CalcularBienes(Calcular):
  def iniciar(same):
    super().iniciar()
    same.tiene_inter = 0
    same.tiene_tv = 0
    same.tiene_compu = 0
    same.tiene_lavadora = 0
    same.tiene_micro = 0
  def cal_bienes(same, tiene_inter , tiene_tv, tiene_compu, tiene_lavadora, tiene_micro):
    if tiene_inter == "Si":
      same.tiene_inter = 1 
    else :
      same.tiene_inter = 0   
    if tiene_tv == "Si":
      same.tiene_tv = 1   
    else :
      same.tiene_tv = 0  
    if tiene_compu == "Si":
      same.tiene_compu = 1
    else :
      same.tiene_compu = 0     
    if tiene_lavadora == "Si":
      same.tiene_lavadora = 1
    else :
      same.tiene_lavadora = 0    
    if tiene_micro == "Si":
      same.tiene_micro = 1
    else :
      same.tiene_micro = 0     
  def reiniciar(same):
      same.tiene_inter= 0
      same.tiene_tv = 0
      same.tiene_compu = 0
      same.tiene_lavadora = 0
      same.tiene_micro = 0
  def cal(same):
      return (same.tiene_inter + same.tiene_tv + same.tiene_compu + same.tiene_lavadora + same.tiene_micro)/5
def cal_total_alpha(num):
     num += (40/2)/(100)
     return num
class CalcularAlimentacion(Calcular):
  def iniciar(same):
    super().iniciar()
    same.alimentacion = 0
  def cal_alimentacion(same,vegetales,carne,leche):
    a = [vegetales,carne,leche]
    bien = a.count('Todo o casi todos los días') + a.count('3 a 5 veces por semana')
    regular = a.count('1 o 2 veces por semana')
    if bien > 2:
      same.alimentacion = 1
    elif (bien>0 and regular>1) or regular == 3 :
      same.alimentacion = 0.5
    else: 
      same.alimentacion = 0
  def reiniciar(same):
      same.alimentacion = 0
  def cal(same):
      return (same.alimentacion)
class CalcularRendimiento(Calcular):
  def iniciar(same):
    super().iniciar()
    same.rendimiento= 0
  def cal_rendimiento(same,rendimiento):
    a = int(rendimiento)
    if a > 350:
      same.rendimiento = 1
    elif a > 250:
      same.redimiento = 0.5
    else:
      same.rendimiento = 0
  def reiniciar(same):
      same.rendimiento=0
  def cal(same):
      return (same.rendimiento)
  



    
```

```python
resul_estrato = CalcularEstrato()
resul_sit_eco = CalcularSituEconomica()
resul_per_hogar = CalcularPersonasHogar()
resul_edu_padre = CalcularEduPadre()
resul_edu_madre = CalcularEduMadre()
resul_bienes = CalcularBienes()
resul_alimentacion = CalcularAlimentacion()
resul_rendimiento = CalcularRendimiento()
```

```python
df["RESULT_ESTRATO"] = np.zeros(df.shape[0])
df["RESULT_SIT_ECO"] = np.zeros(df.shape[0])
df["RESULT_PER_HOGAR"] = np.zeros(df.shape[0])
df["RESULT_EDU_PADRE"] = np.zeros(df.shape[0])
df["RESULT_EDU_MADRE"] = np.zeros(df.shape[0])
df["RESULT_BIENES"] = np.zeros(df.shape[0])
df["RESULT_ALIMEN"] = np.zeros(df.shape[0])
df["RESULT_RENDIMI"] = np.zeros(df.shape[0])
```

```python
for index in df.index:
  resul_estrato.cal_estrato(df["FAMI_ESTRATOVIVIENDA"][index])
  resul_sit_eco.cal_sit_eco(df["FAMI_SITUACIONECONOMICA"][index])
  resul_per_hogar.cal_per_hogar(df["FAMI_PERSONASHOGAR"][index],df["FAMI_SITUACIONECONOMICA"][index])
  resul_edu_padre.cal_edu_padre(df["FAMI_EDUCACIONPADRE"][index])
  resul_edu_madre.cal_edu_madre(df["FAMI_EDUCACIONMADRE"][index])
  resul_bienes.cal_bienes(df["FAMI_TIENEINTERNET"][index],df["FAMI_TIENESERVICIOTV"][index],df["FAMI_TIENECOMPUTADOR"][index],df["FAMI_TIENELAVADORA"][index],df["FAMI_TIENEHORNOMICROOGAS"][index])
  resul_alimentacion.cal_alimentacion(df["FAMI_COMECEREALFRUTOSLEGUMBRE"][index],df["FAMI_COMECARNEPESCADOHUEVO"][index],df["FAMI_COMELECHEDERIVADOS"][index])
  resul_rendimiento.cal_rendimiento(df["PUNT_GLOBAL"][index])
  df.at[index,"RESULT_ESTRATO"] = resul_estrato.cal()
  df.at[index,"RESULT_SIT_ECO"] = resul_sit_eco.cal()
  df.at[index,"RESULT_PER_HOGAR"] = resul_per_hogar.cal()
  df.at[index,"RESULT_EDU_PADRE"] =resul_edu_padre.cal()
  df.at[index,"RESULT_EDU_MADRE"] = resul_edu_madre.cal()
  df.at[index,"RESULT_BIENES"] = resul_bienes.cal()
  df.at[index,"RESULT_ALIMEN"] = resul_alimentacion.cal()
  df.at[index,"RESULT_RENDIMI"] = resul_rendimiento.cal()
 
 
```
